```python
no = []
title = []
year = []
runtime= []
genre = []
certificate=[]
rating = []
metascore = []
description = []
votes = []
gross = []
director = []
cast = []
```

```python
import pandas as pd
from bs4 import BeautifulSoup as soup
import urllib.request
```

```python
url = "https://www.imdb.com/search/title/?groups=top_250&sort=user_rating"

page = urllib.request.urlopen(url)
soup = soup(page,"html.parser")


container = soup.find_all("div",attrs={"class":"lister-item-content"})
for i in soup.find_all("span",attrs={"class":"lister-item-index unbold text-primary"}):
    no.append(int(i.text.strip(".")))
    
    

for i in soup.find_all("h3",attrs={"class":"lister-item-header"}):
    title.append(i.a.text.strip())
    

for i in soup.find_all("span",attrs={"class":"lister-item-year text-muted unbold"}):
    year.append(i.text.strip())
    

for i in soup.find_all("span",attrs={"class":"runtime"}):
    runtime.append(i.text.strip(" min "))
    

for i in soup.find_all("span",attrs={"class":"genre"}):
    genre.append(i.text.strip(" \n "))
    

for i in range(0,len(container)):
    if container[i-1].find("span",attrs={"class":"certificate"})==None:
        certificate.append(None) 
    else:
        certificate.append(container[i-1].find("span",attrs={"class":"certificate"}).text)
        

for i in soup.find_all("div",attrs={"class":"inline-block ratings-imdb-rating"}):
    rating.append(float(i.text.strip()))
    

for i in range(0,len(container)):
    if container[i].find("div",attrs={"class":"inline-block ratings-metascore"})==None:
        metascore.append(None)
    else:
        metascore.append(container[i].find("div",attrs={"class":"inline-block ratings-metascore"}).text.replace("\n","").strip("   Metascore   "))
        

script = soup.find_all("p",attrs={"class":"text-muted"})
for i in range(0,len(container)):
    description.append(script[i*2-1].text.strip())



for i in range(0,len(container)):
    
    votgros = container[i].find("p", attrs={"class": "sort-num_votes-visible"})
    split = votgros.text.split("|")
    if len(split)==2:
        votes.append(split[0].replace("Votes:","").replace("\n","").strip(" "))
        gross.append(split[1].replace("Gross:","").replace("\n","").strip(" "))
    else :
        votes.append(split[0].replace("Votes:","").replace("\n","").strip(" "))
        gross.append(None)


for i in range(0,len(container)):
    p_tag = container[i].find_all("p")
    split = p_tag[2].text.split("|")
    director.append(split[0].strip(" \n ").replace("\n","").replace("Directors:","").replace("Director:",""))
    cast.append(split[1].strip(" \n ").replace("Stars:","").replace("\n",""))
```

```python
imdb = pd.DataFrame({"No":no,"Title":title,"Year":year,"Runtime":runtime,"Genre":genre,"Rating":rating,"Metascore":metascore,"Certificate":certificate,"Description":description,"Votes":votes,"Gross":gross,"Director":director,"Cast":cast})
f = list(imdb.columns)
imdb = imdb[['No','Title','Year','Certificate','Runtime','Genre','Rating','Metascore','Description','Director','Cast','Votes','Gross',]]
    
imdb 
```

```python
imdb.to_csv("imdb top 250.csv")
```
